```python
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
print(datetime.now())
```

```python
df = pd.read_csv('../data/raw/transactions.csv')

df.head()

from datetime import datetime
print(datetime.now())
```

```python
df.info()
df.describe()

from datetime import datetime
print(datetime.now())
```

```python
df['valor'] = df['valor'].astype(float)
df['data'] = pd.to_datetime(df['data'])

df['ano'] = df['data'].dt.year
df['mes'] = df['data'].dt.month

df

from datetime import datetime
print(datetime.now())
```

```python
df_grouped = df.groupby(['tipo'])['valor'].sum().reset_index()

df_grouped

from datetime import datetime
print(datetime.now())
```

```python
df_grouped.plot(kind='bar', x='tipo', y='valor')

plt.title("Total por tipo de transação")

plt.show()

from datetime import datetime
print(datetime.now())
```
